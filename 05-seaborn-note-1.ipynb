### 1.1. `distplot()` -> `displot()`

`sns.distplot()` allows us to see the distribution of univariate set of observation (re: one variable).

It gets us the histogram and a solid line called **KDE (Kernel Density Estimation)**, a non-parametric way to estimate the probability density function of a random variable.

To remove KDE line, simply add argument `kde=False`

### 1.2. `displot()`

`sns.displot()` is a figure-level interface/function which, as the name suggests, gives us a distribution plot with flexible control over the figure properties.

displot() has a histogram style by default, given by the argument `kind='hist'` (in the demonstration below, I typed this argument. But in the real case, you don't need to type it since it's the default value).

It can plot the KDE curve only by setting `kind='kde'`. Alternatively, you can superimpose the KDE instead by giving the argument `kde=True`.

It can also plot the ECDF only with `kind='ecdf'`.

See [full documentation here](https://seaborn.pydata.org/generated/seaborn.displot.html#seaborn.displot).

```python
sns.displot(tips['total_bill'],kind="hist",kde=True)
```

#### 1.2.1. `histplot()`

`sns.histplot()`, as the name suggests, gives us univariate or bivariate histograms to show distributions of datasets.

See the [documentation here](https://seaborn.pydata.org/generated/seaborn.histplot.html#seaborn.histplot).

```python
sns.histplot(tips['total_bill'])
```

#### 1.2.2. `kdeplot()`

`sns.kdeplot()`, as the name suggests, gives us univariate or bivariate distributions using kernel density estimation.

A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, analagous to a histogram. KDE represents the data using a continuous probability density curve in one or more dimensions.

See the [documentation here](https://seaborn.pydata.org/generated/seaborn.kdeplot.html#seaborn.kdeplot).

```python
sns.kdeplot(tips['total_bill'])
```

#### 1.2.3. `ecdfplot()`

`sns.ecdfplot()`, as the name suggests, gives us the Empirical Cumulative Distribution Functions.

An ECDF represents the proportion or count of observations falling below each unique value in a dataset. Compared to a histogram or density plot, it has the advantage that **each observation is visualized directly**, meaning that there are no binning or smoothing parameters that need to be adjusted. It also aids direct comparisons between multiple distributions. **A downside** is that the relationship between the appearance of the plot and the basic properties of the distribution (such as its central tendency, variance, and the presence of any bimodality) may not be as intuitive.

See the [documentation here](https://seaborn.pydata.org/generated/seaborn.ecdfplot.html#seaborn.ecdfplot).

```python
sns.ecdfplot(tips['total_bill'])
```

### 1.3. `jointplot()`

`sns.jointplot()` allows us to match-up two distribution plot for bivariate data (two variables) to compare or observe.

Common Syntax: `sns.jointplot(x=column_name_1,y=column_name_2,data=dataframe,kind="{string}")`

`kind` = { “scatter” | “kde” | “hist” | “hex” | “reg” | “resid” }. By default, `kind` is set to `'scatter'`

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

### 1.4. `pairplot()`

`sns.pairplot()`, as the name suggests, is going to plot pairwise relationships across an entire dataframe, at least for the numerical column. It supports a color hue argument for categorical columns.

It is essentially run the `jointplot()` function for every single possible combination of the numerical columns in the dataframe, and is a nice way to quickly visualize your data.

You can essentially say that `pairplot()` is a **very useful tool for data science**, such as for data exploratory analysis (EDA) using big data.

Common syntax:

`sns.pairplot(dataframe, hue=categorical_column_name,palette=colormap(str))`

`hue` $-$ categorical column name of the dataframe. It usually, but not limited to, contains strings that categorize the entire data.

`palette` $-$ gives a customized coloring based on the category and also return the legend for it. It uses a colormap string argument from matplotlib (or also MATLAB, if you're more familiar with it)

```python
sns.pairplot(tips,hue='sex')
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

### 1.5. `rugplot()`

`sns.rugplot()` plots marginal distributions by drawing ticks along the x and y axes for a single variable.

This function is intended to complement other plots by showing the location of individual observations in an unobstrusive way, and can be used to co-observe the KDE plot.

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.histplot(tips['total_bill'],kde=True)
```

Rugplot() can be used to explain the concept of KDE plot, i.e. how to build the KDE.

In short, the rugplot is the actual distribution point(s). Then we make a normal (Gaussian) distribution for each point. The sum of those Gaussian distributions is what makes our KDE.

### 1.6. KDE Plot $-$ Explained

kdeplots are [Kernel Density Estimation plots](http://en.wikipedia.org/wiki/Kernel_density_estimation#Practical_estimation_of_the_bandwidth). These KDE plots replace every single observation with a Gaussian (Normal) distribution centered around that value.

For example (copied from the lecture's module):

```python
# Don't worry about understanding this code!
# It's just for the diagram below
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Create dataset
dataset = np.random.randn(25)

# Create another rugplot
sns.rugplot(dataset);

# Set up the x-axis for the plot
x_min = dataset.min() - 2
x_max = dataset.max() + 2

# 100 equally spaced points from x_min to x_max
x_axis = np.linspace(x_min,x_max,100)

# Set up the bandwidth, for info on this:
url = 'http://en.wikipedia.org/wiki/Kernel_density_estimation#Practical_estimation_of_the_bandwidth'

bandwidth = ((4*dataset.std()**5)/(3*len(dataset)))**.2


# Create an empty kernel list
kernel_list = []

# Plot each basis function
for data_point in dataset:
    
    # Create a kernel for each point and append to list
    kernel = stats.norm(data_point,bandwidth).pdf(x_axis)
    kernel_list.append(kernel)
    
    # Scale for plotting
    kernel = kernel / kernel.max()
    kernel = kernel * .4
    plt.plot(x_axis,kernel,color = 'grey',alpha=0.5)

plt.ylim(0,1)
```

To get the kde plot, we can sum these basis functions.

```python
# Plot the sum of the basis function
sum_of_kde = np.sum(kernel_list,axis=0)

# Plot figure
fig = plt.plot(x_axis,sum_of_kde,color='indianred')

# Add the initial rugplot
sns.rugplot(dataset,c = 'indianred')

# Get rid of y-tick marks
plt.yticks([])

# Set title
plt.suptitle("Sum of the Basis Functions")
```
