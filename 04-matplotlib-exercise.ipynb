## Data

````python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
````

````python
import matplotlib.pyplot as plt
%matplotlib inline
````

Use plt.show() at the end of the plotting code if you're not using jupyter environment.

## Exercise 1

1. Create a figure object called **fig** using *plt.figure()*, then use *add_axes* as **ax** to add an axis to the *[0,0,1,1]* canvas.
2. Plot (x,y) on that axes and set the labels and title.

````python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Title')
# alternatively, plt. can be used instead of ax.
````

## Exercise 2

1. Create a figure object and put two axes on it, **ax1** and **ax2**, located at [0,0,1,1] and [0.2,0.5,.2,.2], respectively.
2. Plot (x,y) on both axes and call your figure object to show it.

````python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])

ax1.plot(x,y)
ax1.set_xlabel('x')
ax1.set_ylabel('y')

ax2.plot(x,y)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
````

## Exercise 3

1. Create a plot by adding two axes to a figure object at ax1 = [0,0,1,1] and ax2 = [0.2,0.5,.4,.4].
2. Plot (x,z) into ax1 and (x,y) into ax2.
3. Notice the axis limit, re: ax1 = [0 100 0 10000], ax2 = [20 22 30 50] (note: this is MATLAB format to read axis)

````python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])

ax1.plot(x,z,'b')
ax1.set_xlim([0,100])
ax1.set_ylim([0,10000])
ax1.set_xlabel('X')
ax1.set_ylabel('Z')

ax2.plot(x,y,'r')
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('Zoom')
````

## Exercise 4

1. Use `plt.subplots(nrows=1,ncols=2)` to create a plot.
2. Plot (x,y) and (x,z) on the axes. Play around with the linewidth and style!
3. In a new cell (if you use jupyter system), recreate the previous plot and add a `figsize` argument. Put any size you want!

````python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y,ls='--',lw=2,color='blue')
axes[0].set_xlim([0,100])
axes[0].set_ylim([0,200])

axes[1].plot(x,z,linestyle='-', linewidth=2.5, color='black') # linestyle can be skipped since '-' is the default style
axes[1].set_xlim([0,100])
axes[1].set_ylim([0,10000])
````

````python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(9,2))

axes[0].plot(x,y,ls='-',lw=2,color='blue')
axes[0].set_xlim([0,100])
axes[0].set_ylim([0,200])

axes[1].plot(x,z,linestyle='--', linewidth=2.5, color='red') # linestyle can be skipped since '-' is the default style
axes[1].set_xlim([0,100])
axes[1].set_ylim([0,10000])

plt.tight_layout()
````
